```python
# SCENARIO 1: E-commerce Sales Data Analysis
print("swathi-24BAD122")

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload file
uploaded = files.upload()

# Get uploaded filename automatically
file_name = list(uploaded.keys())[0]

# Step 2: Read file (handles both CSV and ZIP)
try:
    data = pd.read_csv(file_name, encoding='ISO-8859-1')
except:
    data = pd.read_csv(file_name, compression='zip', encoding='ISO-8859-1')

print("\nFile Loaded Successfully âœ…")

# ----------------------------
# Dataset Inspection
# ----------------------------

print("\nFirst 5 Rows:")
print(data.head())

print("\nLast 5 Rows:")
print(data.tail())

print("\nDataset Info:")
print(data.info())

print("\nStatistical Summary:")
print(data.describe())

print("\nMissing Values:")
print(data.isnull().sum())

# ----------------------------
# Sales Analysis
# ----------------------------

# Remove null descriptions
data = data.dropna(subset=["Description"])

# Group by product and sum quantity
sales_per_product = (
    data.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# ----------------------------
# Bar Chart
# ----------------------------

plt.figure(figsize=(12,6))
sales_per_product.plot(kind='bar')
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()

# ----------------------------
# Line Chart
# ----------------------------

plt.figure(figsize=(12,6))
sales_per_product.plot(kind='line', marker='o')
plt.title("Sales Trend of Top 10 Products")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```
